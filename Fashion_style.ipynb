```python
# MODEL TRSINING USING CNN
# involves convo,pooling,flsttening,tehn ann model/ the neural netword
#first 3 for spatial idetity retetion of the pixels
```

```python
# create training set, resukt set
X_train = training[:,1:]/255 # /255 for normalization
y_train = training[:,0]
```

```python
X_test = testing[:,1:]/255 # /255 for normalization
y_test = testing[:,0]
```

```python
# to avoid overfitting and generalization in cnn use validation to make sure of it
from sklearn.model_selection import train_test_split
```

```python
X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, test_size = 0.2, random_state = 12345)
#validate is like test only
```

```python
#cnn expects the input to be in the form of an image
#reshape the pixels of the X data to form an image of 28x28x1
X_train = X_train.reshape(X_train.shape[0], *(28, 28, 1))
X_test = X_test.reshape(X_test.shape[0], *(28, 28, 1))
X_validate = X_validate.reshape(X_validate.shape[0], *(28, 28, 1))
#on this apply convo and so on
# each image is a gray scale image in this formate
```

```python
X_train.shape
```

```python
X_validate.shape
```

```python
# we already have images in pixels / numbers
import keras
```

```python
# get the layers to convo, pool,flaten etc
from keras.models import Sequential # to build layers in a sequence
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
# we are not using tensor flow here , might reduce edd maybe not sure but insted v r using keeras libs  directly
```

```python
#build cnn in seqence
cnn_model = Sequential()
```

```python
#convolution and activation
cnn_model.add(Conv2D(32,3,3,input_shape=(28,28,1),activation = 'relu')) #num of kernels,dimension
```

```python
#pooling --> for simplicity
cnn_model.add(MaxPooling2D(pool_size = (2,2))) # the size of pooling function
```

```python
#flatten it --> so that simplified data is in a araayish format
cnn_model.add(Flatten())
```

```python
# dense application ie building of rest of the layers including output
cnn_model.add(Dense(units = 32, activation = 'relu'))
```

```python
cnn_model.add(Dense(units = 10, activation = 'sigmoid'))
```

```python
# use adam optimizer , loss entrophy function
cnn_model.compile(loss ='sparse_categorical_crossentropy', optimizer=Adam(lr=0.001),metrics =['accuracy']) # this loss fnc as many categories
# loss to know diff b/w op pred and actual
# optimizerr to update weights
```

```python
# fit our model --> train it ie
epochs = 50 #num of times to update weight
history = cnn_model.fit(X_train,
                        y_train,
                        batch_size = 512,
                        epochs = 50,
                        verbose = 1,
                        validation_data = (X_validate, y_validate)) # last part for cross validation

# i assume the number of dataset reduced coz of g collab
```

```python
#evaluation
```

```python
#done directly using the evaluate function
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy : {:.3f}'.format(evaluation[1]))
# during testing the accuracy dropped a bit
```

```python
predictions = cnn_model.predict(X_test)

# Get the predicted class labels
predicted_classes = np.argmax(predictions, axis=1)
```

```python
predicted_classes
```

```python
L = 5
W = 5
fig, axes = plt.subplots(L, W, figsize = (12,12))
axes = axes.ravel() #

for i in np.arange(0, L * W):
    axes[i].imshow(X_test[i].reshape(28,28))# conver tthe pixel data to image
    axes[i].set_title("Prediction Class = {:0.1f}\n True Class = {:0.1f}".format(predicted_classes[i], y_test[i])) # this is used to get the predicteed and actual class
    axes[i].axis('off')
plt.subplots_adjust(wspace=0.5)

# here we visualize for a few data and compare the pred and actual
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_classes)
plt.figure(figsize = (14,10))
sns.heatmap(cm, annot=True)
# Sum the diagonal element to get the total true correct values

# good number of misses and hits here
```

```python
from sklearn.metrics import classification_report

num_classes = 10
target_names = ["Class {}".format(i) for i in range(num_classes)]

print(classification_report(y_test, predicted_classes, target_names = target_names))
```
